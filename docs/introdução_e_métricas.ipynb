# Introduction

This project employed `MEAMT` (Multiobjective Evolutionary Algorithm with many tables) for solving benchmarks problems in hope to demonstrate the algorithm potential for multiobjective optimization.

The algorithm works by isolating combinations of objectives in subpopulations called tables such that each table contains the best individuals in the population for any combination, this is done so that individuals that might have relevance for optimizing a certain objective or a combination of those are not discarded due to elitism and the algorithm can remain diverse enough to escape local minimals.

For assessement of the results the pareto front was plotted in 3d and compared to the know solution of each benchmark, some well establishment metrics for MOEAs were also assessed for a more complete analysis.

# Perfomance Metrics

## IGD+
IGD+ (Inverted Generational Distance Plus), proposed as an evolution of IGD [3], works by indicating how far the current set is from the solution. The closer it is to zero, the better.

This metric exists to solve the main problem with the IGD. Unlike IGD, IGD+ satisfies Weak Pareto Compliance, which guarantees that if a solution set $A$ dominates a set $B$, the metric will not contradict this relationship (i.e., $IGD^+(A) \le IGD^+(B)$).

Mathematically, IGD+ calculates the average (or root-mean-power) distance from each point in the reference set ($Z$) to its nearest neighbor in the approximation set ($A$), using a one-sided distance calculation:
\begin{equation}
IGD^+(A, Z) = \frac{1}{|Z|} \left( \sum_{j=1}^{|Z|} d^+_j(z, a)^p \right)^{1/p}
\end{equation}

where $a \in A \subset R^m, z \in Z \subset R^m$, $A$ is the Pareto set approximation and $Z$ is the reference set. $d^+(a, z)$ is defined as

\begin{equation}
d^+(z, a) = \sqrt{ \sum_{i=1}^{m} (\max(a_i - z_i, 0))^2 }
\end{equation}


## Hypervolume (HV)
The **Hypervolume (HV)**, originally proposed as the *S-metric* or *Size of the space covered* [1], is one of the most widely used metrics in multi-objective optimization. Its popularity stems from the fact that it is a unary metric capable of simultaneously capturing two crucial characteristics of a solution set:

1.  **Convergence:** The proximity of the solutions to the true Pareto Front.
2.  **Diversity:** The spread and distribution of solutions within the objective space.

Mathematically, Hypervolume calculates the volume of the objective space that is "dominated" by the obtained solution set ($S$), bounded by a reference point ($r$):

$$
HV(S, r) = \text{Vol} \left( \bigcup_{x \in S} [f_1(x), r_1] \times \dots \times [f_m(x), r_m] \right)
$$

The larger the Hypervolume value, the better the quality of the solution set.

# Benchmarks

## DTLZ1
Part of the DTLZ suite by Deb-Thiele-Laumanns-Zitzler, this test evaluate an algorithm perfomance by testing its converge capacity. Its function creates thousands of local Pareto Fronts that works as "traps". The algorithm needs to have a good global exploration to not get stuck in local optimum.

**Mathematical Definition**

The problem is defined by minimizing M objective functions (3 in our test), through otimization of a vector with n decision variables $\mathbf{x} = (x_1, x_2, \dots, x_n)^T$ where each variable is limited by the interval  $x_i \in [0, 1]$.The total number of variables is $n = M + k - 1$. For DTLZ1, the recommended number of distance variables is typically $k = 5$.

The function penality/distance $g(\mathbf{x}_M)$, causes great multimodality, this is, causes many local optima,and is defined as:
$$g(\mathbf{x}_M) = 100 \left[ k + \sum_{i=M}^{n} \left( (x_i - 0.5)^2 - \cos(20\pi(x_i - 0.5)) \right) \right]$$

The M objectives to be minimized are formulated as:

$$f_1(\mathbf{x}) = \frac{1}{2} x_1 x_2 \cdots x_{M-1} (1 + g(\mathbf{x}_M))
$$
$$f_2(\mathbf{x}) = \frac{1}{2} x_1 x_2 \cdots (1 - x_{M-1}) (1 + g(\mathbf{x}_M))
$$
$$\vdots$$
$$f_{M-1}(\mathbf{x}) = \frac{1}{2} x_1 (1 - x_2) (1 + g(\mathbf{x}_M))$$
$$f_M(\mathbf{x}) = \frac{1}{2} (1 - x_1) (1 + g(\mathbf{x}_M))$$

## DTLZ2
The DTLZ2 problem is another fundamental benchmark from the Deb-Thiele-Laumanns-Zitzler test suite. While DTLZ1 is designed to test an algorithm's ability to converge through a severe, multimodal landscape, DTLZ2 is designed to evaluate how well an algorithm can maintain a diverse set of solutions across a continuous, curved surface while being unimodal.

**Mathematical Definition**

The problem is defined to minimize $M$ objective functions by optimizing a decision vector $\mathbf{x} = (x_1, x_2, \dots, x_n)^T$, where each variable is bounded within $x_i \in [0, 1]$.The total number of variables is $n = M + k - 1$. For DTLZ2, the recommended number of distance variables is typically $k = 10$. The variables are split into position variables ($x_1$ to $x_{M-1}$) and distance variables ($\mathbf{x}_M$, ranging from $x_M$ to $x_n$).The distance function $g(\mathbf{x}_M)$ for DTLZ2 is a simple spherical function, defined as:$$g(\mathbf{x}_M) = \sum_{x_i \in \mathbf{x}_M} (x_i - 0.5)^2$$The $M$ objectives to be minimized are formulated using sine and cosine mappings to create the spherical front:$$f_1(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1 \pi}{2}\right) \cos\left(\frac{x_2 \pi}{2}\right) \cdots \cos\left(\frac{x_{M-2} \pi}{2}\right) \cos\left(\frac{x_{M-1} \pi}{2}\right)$$
$$f_2(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1 \pi}{2}\right) \cos\left(\frac{x_2 \pi}{2}\right) \cdots \cos\left(\frac{x_{M-2} \pi}{2}\right) \sin\left(\frac{x_{M-1} \pi}{2}\right)$$
$$f_3(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1 \pi}{2}\right) \cos\left(\frac{x_2 \pi}{2}\right) \cdots \sin\left(\frac{x_{M-2} \pi}{2}\right)$$
$$\vdots$$
$$f_M(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \sin\left(\frac{x_1 \pi}{2}\right)$$

## DTLZ3
The DTLZ3 test function is a scalable, multi-objective optimization problem introduced by Deb et al. (2002). It is specifically designed to evaluate a Multi-Objective Evolutionary Algorithm's (MOEA) ability to converge to the global Pareto-optimal front in the presence of massive multi-modality[2].

**Mathematical Definition**

The objective functions to be minimized are defined as follows:$$\begin{aligned}
\min f_1(\mathbf{x}) &= (1 + g(\mathbf{x}_M)) \cos(x_1\pi/2) \dots \cos(x_{M-1}\pi/2) \\
\min f_2(\mathbf{x}) &= (1 + g(\mathbf{x}_M)) \cos(x_1\pi/2) \dots \sin(x_{M-1}\pi/2) \\
\vdots \\
\min f_M(\mathbf{x}) &= (1 + g(\mathbf{x}_M)) \sin(x_1\pi/2) \\
\end{aligned}$$The ruggedness of the landscape is governed by the $g(\mathbf{x}_M)$ function, which uses a Rastrigin-like transformation:$$g(\mathbf{x}_M) = 100 \left[ |x_M| + \sum_{x_i \in \mathbf{x}_M} ((x_i - 0.5)^2 - \cos(20\pi(x_i - 0.5))) \right]$$

## DTLZ4
The DTLZ4 problem is a direct modification of DTLZ2. While DTLZ2 evaluates an algorithm's basic ability to find and cover a continuous spherical front, DTLZ4 is specifically designed to test an algorithm's ability to maintain a good distribution of solutions when the search space is highly biased.

**Mathematical Definition**

As in the previous problems of the suite, the goal is to minimize $M$ functions using a vector of $n$ variables $\mathbf{x} = (x_1, x_2, \dots, x_n)^T$, where each variable is bounded within $x_i \in [0, 1]$. The total number of variables is $n = M + k - 1$, with $k = 10$ typically adopted for the distance variables.The distance function $g(\mathbf{x}_M)$ remains unchanged from DTLZ2, being a simple spherical function:$$g(\mathbf{x}_M) = \sum_{x_i \in \mathbf{x}_M} (x_i - 0.5)^2$$The structural difference lies in the formulation of the $M$ objectives, which apply the exponent $\alpha$ to all position variables ($x_1$ to $x_{M-1}$) inside the trigonometric functions:$$f_1(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1^\alpha \pi}{2}\right) \cos\left(\frac{x_2^\alpha \pi}{2}\right) \cdots \cos\left(\frac{x_{M-2}^\alpha \pi}{2}\right) \cos\left(\frac{x_{M-1}^\alpha \pi}{2}\right)$$
$$f_2(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1^\alpha \pi}{2}\right) \cos\left(\frac{x_2^\alpha \pi}{2}\right) \cdots \cos\left(\frac{x_{M-2}^\alpha \pi}{2}\right) \sin\left(\frac{x_{M-1}^\alpha \pi}{2}\right)$$
$$f_3(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \cos\left(\frac{x_1^\alpha \pi}{2}\right) \cos\left(\frac{x_2^\alpha \pi}{2}\right) \cdots \sin\left(\frac{x_{M-2}^\alpha \pi}{2}\right)$$
$$\vdots$$
$$f_M(\mathbf{x}) = (1 + g(\mathbf{x}_M)) \sin\left(\frac{x_1^\alpha \pi}{2}\right)$$

## ZDT1
The ZDT1 problem is the foundational benchmark from the Zitzler-Deb-Thiele (ZDT) test suite. Unlike the DTLZ suite, which scales to any number of objectives ($M$), the standard ZDT problems are strictly designed for bi-objective optimization ($M = 2$). ZDT1 serves as the most basic baseline to test an algorithm's ability to converge to a simple, continuous Pareto front being unimodal and serving as a "easy" problem for testing new algoritms.

**Matematical Definition**

The problem is defined to minimize two objective functions, $f_1$ and $f_2$, by optimizing a decision vector of $n$ variables $\mathbf{x} = (x_1, x_2, \dots, x_n)^T$, where each variable is bounded within $x_i \in [0, 1]$. Typically, for ZDT1, the total number of decision variables is set to $n = 30$.The ZDT problems are generally structured using three functions: $f_1$ (which primarily controls the first objective), $g$ (the distance/convergence function), and $h$ (the shape function that defines the Pareto front's geometry).For ZDT1, these functions are defined as:$$f_1(\mathbf{x}) = x_1$$
$$g(\mathbf{x}) = 1 + \frac{9}{n - 1} \sum_{i=2}^{n} x_i$$
$$h(f_1, g) = 1 - \sqrt{\frac{f_1}{g}}$$
The second objective function is then formulated by multiplying the distance and shape functions:$$f_2(\mathbf{x}) = g(\mathbf{x}) \cdot h(f_1, g)$$

## ZDT3

The ZDT3 is a popular continuous test problem from the Zitzler-Deb-Thiele (ZDT) suite, which is extensively used to evaluate the performance of Multi-Objective Evolutionary Algorithms (MOEAs).

What makes ZDT3 stand out from the rest of the suite is the introduction of a disconnected Pareto optimal front. The presence of a sine wave in its formulation causes the optimal solutions to be distributed across several separate, non-contiguous convex regions.

**Mathematical Definition**

The ZDT3 problem is defined to minimize two objective functions, $f_1(x)$ and $f_2(x)$. The function is split into three parts: the first objective $f_1$, a diversity function $g(x)$, and a shape function $h(f_1, g)$.1. First Objective:$$f_1(x) = x_1$$2. Diversity Function:$$g(x) = 1 + \frac{9}{n-1} \sum_{i=2}^n x_i$$3. Second Objective:The second objective incorporates the shape function $h$, which contains the sine term responsible for the disconnected front:$$f_2(x) = g(x) \left( 1 - \sqrt{\frac{f_1(x)}{g(x)}} - \frac{f_1(x)}{g(x)} \sin(10 \pi f_1(x)) \right)$$
